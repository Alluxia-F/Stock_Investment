```python
sp500_comp=pd.read_excel('sp500_company_list_weights.xlsx')
```

### Load the file
### Generate the file including companies list, weights and sector

```python
sp500_comp.head()
```

```python
sp500_list=sp500_comp.drop(['Price','Change'],axis=1)
sp500_list.head()
```

```python
sp500_list.shape
```

```python
sp500_sector=pd.read_excel('sp500_wiki.xlsx')
sp500_sector.head(10)
```

```python
sp500_sector.shape
```

```python
sp500_company_sector=pd.merge(sp500_list,sp500_sector,on=['Symbol'])
sp500_company_sector.head()
```

```python
sp500_company_sector.info()
```

```python
sp500_company_sector.drop('Company_y',axis=1,inplace=True)
```

```python
sp500_final=sp500_company_sector.rename(columns={'Company_x':'Company'})
sp500_final.tail(10)
```

```python
sp500_final.to_csv('sp500_companies_weights_sector.csv')
sp500_final.head()
```

### Analysis companies and weights by sector

```python
sp500_final.groupby('Sector')['Company','Symbol'].count()
```

```python
sp500_final.groupby('Sector')['Weight'].sum().round(1)
```

```python
columns=list(set(sp500_final['Sector']))
columns
```

#### Generate the output file listing companies in each sector

```python
sector_list=['Utilities','Financials','Communication Services','Information Technology','Materials','Real Estate',\
             'Energy','Consumer Staples','Industrials','Consumer Discretionary','Health Care']

df_list=['utilities','financials','communication_services','information_technology','materials','real_estate',\
             'energy','consumer_staples','industrials','consumer_discretionary','health_care']
len(sector_list)
```

```python
for i in range(11):
    df_list[i]=sp500_final.groupby('Sector').get_group(sector_list[i]).sort_values(by='Weight',ascending=False)
    print(sector_list[i]+' has: '+str(df_list[i].shape[0])+' Companies!\n')
    df_list[i].to_csv('sp500_sector/'+sector_list[i]+'.csv')
print('File generated!')
```
